```python
res=requests.get("https://codeforces.com/api/user.status?handle="+usr).json()
res
```

```python
usr_data = {}
```

```python
usr_info = requests.get("https://codeforces.com/api/user.info?handles="+usr).json()
usr_info
```

```python
usr_data['rating'] = usr_info['result'][0]['rating']
usr_data['max_rating'] = usr_info['result'][0]['maxRating']
usr_data['cf_handle'] = usr_info['result'][0]['handle']
usr_data['rank'] = usr_info['result'][0]['rank']

usr_data['category_count'] = {}
usr_data['solved_problems'] = []
```

```python
for intel in res['result']:
    if intel['verdict'] == 'OK':
        prlbm = intel['problem']
        usr_data['solved_problems'].append(str(prlbm['contestId'])+str(prlbm['index']))
        for tag in prlbm['tags']:
            if tag not in usr_data['category_count']:
                usr_data['category_count'][tag] = {}
                if prlbm['index'] not in usr_data['category_count'][tag]:
                    usr_data['category_count'][tag][prlbm['index']] = 1
                else:
                    usr_data['category_count'][tag][prlbm['index']]+=1
            else:
                if prlbm['index'] not in usr_data['category_count'][tag]:
                    usr_data['category_count'][tag][prlbm['index']] = 1
                else:
                    usr_data['category_count'][tag][prlbm['index']]+=1
```
